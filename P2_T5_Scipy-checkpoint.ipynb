```python
from scipy import integrate
import numpy as np
```

```python
## When we integrate any function where analytically integrate is not possible,
## Scipy provides functionality to integrate  funtion with numeriacal integration
```

```python
x2=lambda x:x**2
print(integrate.quad(x2,0,4),'with error')# second value is estimated error in integral
print(4**3/3)# Analitical result
```

```python
# int(e^-x)=(-e^-x)
invexp=lambda x:np.exp(-x)
print(integrate.quad(invexp,0,np.inf))
a,b=integrate.quad(invexp,0,np.inf)
print(a)
print(b)
```

```python
## [sin x] from 0 to pi
## sin(x)=int[-cos(x)]
invsin=lambda x:np.sin(x)
print(integrate.quad(invsin,0,np.pi))
```

## Double integeration

```python
# import square root function from math lib
from math import sqrt
# set function f(x)
f=lambda x,y: 64*x*y
# lower limit of second integral 
p=lambda x:0
## upper limit of second integral
q=lambda y:sqrt(1-2*y**2)
# perform db intg
integration=integrate.dblquad(f,0,2/4,p,q)
print(integration)
```

# Optimization

```python
# image processing me is type ka optimization hota hai
```

```python
from scipy import optimize
from matplotlib import pyplot as plt
```

```python
## finding the minimum of scalar function 
def f(x):
    return x**2+10*np.sin(x)

x=np.arange(-10,10,0.1)
plt.plot(x,f(x))
plt.show()
```

```python
## BFGS Algorithm
#### find global minima
print(optimize.fmin_bfgs(f,3))
```

```python
print(optimize.fmin_bfgs(f,0))
```

```python
## basinhopping algo
# minimization value deta, hai start value always 0
```

```python
print(optimize.basinhopping(f,0))
```

## Scipy with images

```python
from scipy import misc
from matplotlib import pyplot as plt
import numpy as np
# get face image from misc package
panda=misc.face()
# plot or show image face
plt.imshow(panda)
plt.show()
```

```python
# image represented in 3 plane(color)
# R G B
```

```python
plt.imshow(panda.reshape(1024,768,3)) # reshape
plt.show()
```

```python
from scipy import ndimage, misc
# rotation function of scipy for image image rotatet 135 degree
panda_rotate=ndimage.rotate(panda,90)
plt.imshow(panda_rotate)
plt.show()
```
